```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
dataset=pd.read_csv("D:/doctor recoomendation/archive (8)/Original_Dataset.csv")
```

```python
dataset.head(100)
```

```python
dataset.shape
```

```python
symno = []
for col in dataset.columns:
    if col != 'Disease':
        symno.append(col)
print(symno)
```

```python
symtoms2=dataset.iloc[:,1:].values.flatten()
```

```python
symtoms2
```

```python
symtoms=list(set(symtoms2))
```

```python
symtoms
```

```python
for symptom in symtoms:
    dataset[symptom] = dataset.iloc[:, 1:].apply(lambda row: int(symptom in row.values), axis=1)
dataset1= dataset.drop(columns=symno)
```

```python
dataset1
```

```python
dataset1=dataset1.loc[:, dataset1.columns.notna()]
```

```python
dataset1
```

```python
df=open("columns.txt","w")
dataset1.columns
for i in dataset1.columns:
    df.write(i+"\n")
df.close()
```

```python
from sklearn.preprocessing import LabelEncoder
var_mod = ['Disease']
le = LabelEncoder()
for i in var_mod:
    dataset1[i] = le.fit_transform(dataset1[i])
```

```python
dataset1
```

```python

X = dataset1.drop(columns="Disease")
y = dataset1['Disease']
```

```python
model=LogisticRegression()
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
model2=RandomForestClassifier()
model2.fit(X,y)
pred_2=model2.predict(X)
model.fit(X,y)
pred = model.predict(X)
accuracy = metrics.accuracy_score(pred,y)
accuracy_2=metrics.accuracy_score(pred_2,y)
print(model,'Accuracy : %s' % '{0:.2%}'.format(accuracy))
print(model2,'Accuracy : %s' % '{0:.2%}'.format(accuracy_2))
```

```python
test_col = []
for col in dataset1.columns:
    if col != 'Disease':
        test_col.append(col)

symtoms_input=[]
print('enter the symptoms and numbef from below')
print(symtoms)

num_inputs = int(input("Enter the number of symptoms you have: "))
for i in range(num_inputs):
    user_input = input("Enter Symptoms #{}: ".format(i+1))
    symtoms_input.append(user_input)
print("Symptoms you have:", symtoms_input)
no_of_symtoms=len(symtoms_input)
```

```python
test_data={}
for column in test_col:
        test_data[column] = 1 if column in symtoms_input else 0
        test_df = pd.DataFrame(test_data, index=[0])
test_df
```

```python
predict_disease=model.predict(test_df)
predict_disease=le.inverse_transform(predict_disease)
print(predict_disease)

```
